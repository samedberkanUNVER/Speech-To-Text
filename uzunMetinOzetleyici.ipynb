```python


import openai
from google.colab import files

# OpenAI API anahtarınızı buraya girin
openai.api_key = 'APEN API KEY'

def read_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

def split_text(text, chunk_size=3000):
    # Metni belirtilen boyutta parçalara bölelim
    return [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

def summarize_text(text):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",  # Veya gpt-4
            messages=[{
                "role": "system",
                "content": "Aşağıdaki metni özetleyin. Sadece Türkçe yaz."
            }, {
                "role": "user",
                "content": text
            }]
        )
        return response['choices'][0]['message']['content']
    except Exception as e:
        print(f"Error summarizing text: {e}")
        return None

def summarize_large_text(file_path):
    full_text = read_file(file_path)
    chunks = split_text(full_text)

    summaries = []
    for chunk in chunks:
        summary = summarize_text(chunk)
        if summary:
            summaries.append(summary)

    # Tüm özetleri birleştirelim
    return " ".join(summaries)

# Dosya yükleme kısmı
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Özetleme işlemi
summarized_text = summarize_large_text(file_name)
print("Summarized Text:")
print(summarized_text)
```
